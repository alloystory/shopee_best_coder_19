```python
import re

regex = re.compile('[^a-zA-Z]')


def remove_unwanted(string):
  res= regex.sub(' ', string)
  return res
```

```python
def splitter(string):
  res = string.split(' ')
  return res
```

```python
spam['alphabetic'] = spam['name'].apply(remove_unwanted)
```

```python
spam['splitted'] = spam['alphabetic'].apply(splitter)
```

```python
spam['splitted'] = spam['splitted'].apply(lambda x: list(filter(None, x)))
```

```python
spam['splitted'] = spam['splitted'].apply(lambda x: [i.lower() for i in x])
```

```python
from fuzzywuzzy import fuzz
```

```python
fuzz.partial_ratio('coat','cost')
```

```python
def threshold_on(extra_dict, alist):
  res_list = []
  for word in alist:
    for matcher in extra_dict:
      score=fuzz.ratio(word, matcher)
      if score > 80:
        res_list.append((word, extra_dict[matcher]))
  return list(set([x[1] for x in res_list]))
```
